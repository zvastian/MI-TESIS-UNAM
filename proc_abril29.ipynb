```python
import pandas as pd

# --------------------------------------------------------------
# 1. Cargar el archivo Parquet (cambia la ruta según tu archivo)
# --------------------------------------------------------------
ruta_parquet = "tesis_vectorizada.parquet"  # <--- MODIFICA AQUÍ
df = pd.read_parquet(ruta_parquet)

# --------------------------------------------------------------
# 2. Identificar columnas que contienen "plantel" (ignorando mayúsculas)
# --------------------------------------------------------------
columnas_plantel = [col for col in df.columns if "plantel" in col.lower()]

if not columnas_plantel:
    print("No se encontraron columnas con 'plantel' en el nombre.")
else:
    print(f"Columnas encontradas: {columnas_plantel}\n")

# --------------------------------------------------------------
# 3. Analizar cada columna
# --------------------------------------------------------------
resumen = []
for col in columnas_plantel:
    # Número total de filas (documentos)
    total_filas = len(df)
    # Valores no nulos
    no_nulos = df[col].notna().sum()
    nulos = total_filas - no_nulos
    # Valores únicos (distintos planteles)
    unicos = df[col].nunique()
    # Ejemplo de algunos valores únicos (primeros 5)
    ejemplos = df[col].dropna().unique()[:5].tolist()
    
    resumen.append({
        "columna": col,
        "total_filas": total_filas,
        "no_nulos": no_nulos,
        "nulos": nulos,
        "%_no_nulos": round(100 * no_nulos / total_filas, 2),
        "unicos": unicos,
        "ejemplos": ejemplos
    })

# --------------------------------------------------------------
# 4. Mostrar resultados y recomendar la mejor columna
# --------------------------------------------------------------
df_resumen = pd.DataFrame(resumen)
print(df_resumen[["columna", "no_nulos", "%_no_nulos", "unicos", "ejemplos"]])
print("\n" + "="*80)

# Criterio: elegir la columna con más valores no nulos.
# Si hay empate, la que tenga más valores únicos (mayor diversidad).
mejor = df_resumen.sort_values(by=["no_nulos", "unicos"], ascending=False).iloc[0]
print(f"RECOMENDACIÓN: La columna '{mejor['columna']}' parece ser la más avanzada porque tiene:\n"
      f"  - {mejor['no_nulos']} valores no nulos ({mejor['%_no_nulos']}% del total)\n"
      f"  - {mejor['unicos']} valores únicos de plantel\n"
      f"  - Ejemplos: {mejor['ejemplos'][:3]} ...")

# Opcional: guardar resumen en CSV
df_resumen.to_csv("analisis_columnas_plantel.csv", index=False, encoding="utf-8")
print("\nResumen guardado en 'analisis_columnas_plantel.csv'")
```

```python
import pandas as pd
import re
import unicodedata
from difflib import get_close_matches

# --------------------------------------------------------------
# 1. Cargar el DataFrame original (con plantel_corregido)
# --------------------------------------------------------------
df = pd.read_parquet("t.parquet")  # o el nombre de tu archivo

# --------------------------------------------------------------
# 2. Cargar el CSV de valores únicos con ejemplos (para construir el mapeo)
# --------------------------------------------------------------
# Asegúrate de que este archivo existe y tiene las columnas 'plantel_corregido', 'freq', 'ejemplo_sec', 'ejemplo_nota'
valores_unicos = pd.read_csv("valores_unicos_con_ejemplos.csv")

# --------------------------------------------------------------
# 3. Función de normalización agresiva
# --------------------------------------------------------------
def normalizar(texto):
    if not isinstance(texto, str):
        return texto
    # Minúsculas
    texto = texto.lower()
    # Eliminar acentos
    texto = unicodedata.normalize('NFKD', texto).encode('ASCII', 'ignore').decode('ASCII')
    # Eliminar puntuación que no sea espacios o letras/números
    texto = re.sub(r'[^\w\s]', '', texto)
    # Eliminar sufijos comunes (institución que otorga el título, etc.)
    sufijos = [
        'institucion que otorga el titulo', 'institucion que otorga el grado',
        'entidad participante', 'programa participante', 'escuela de',
        'facultad de', 'plantel', 'colegio', 'instituto'
    ]
    for suf in sufijos:
        texto = re.sub(rf'\b{suf}\b', '', texto)
    # Eliminar palabras comunes genéricas que no identifican
    palabras_eliminar = ['de', 'la', 'el', 'las', 'los', 'y', 'del', 'para']
    for p in palabras_eliminar:
        texto = re.sub(rf'\b{p}\b', ' ', texto)
    texto = re.sub(r'\s+', ' ', texto).strip()
    return texto

# --------------------------------------------------------------
# 4. Construir un mapeo usando fuzzy matching
# --------------------------------------------------------------
nombres_originales = valores_unicos['plantel_corregido'].tolist()
nombres_normalizados = [normalizar(n) for n in nombres_originales]

# Diccionario para almacenar el mapeo: clave = nombre original, valor = nombre canónico
mapeo = {}

# Procesar grupos de nombres similares
umbral = 0.85  # ajusta según necesidad (0.8 a 0.9)
ya_asignados = set()
for i, orig in enumerate(nombres_originales):
    if orig in ya_asignados:
        continue
    norm_i = nombres_normalizados[i]
    # Buscar todos los índices con normalizados similares a norm_i
    similares = [j for j, norm_j in enumerate(nombres_normalizados)
                 if get_close_matches(norm_j, [norm_i], n=1, cutoff=umbral)]
    # Elegir el mejor nombre canónico del grupo: el más largo (o el que tenga más información)
    grupo = [nombres_originales[j] for j in similares]
    # Criterio: el que tenga mayor longitud (más completo) y no empiece por "none"
    canonicos = [g for g in grupo if g.lower() != 'none']
    if canonicos:
        canonico = max(canonicos, key=len)
    else:
        canonico = grupo[0] if grupo else orig
    # Asignar a todos los del grupo
    for j in similares:
        mapeo[nombres_originales[j]] = canonico
        ya_asignados.add(nombres_originales[j])

# Añadir también los que no se agruparon (como 'none')
for orig in nombres_originales:
    if orig not in mapeo:
        mapeo[orig] = orig if orig.lower() != 'none' else None

# --------------------------------------------------------------
# 5. Limpiar el nombre canónico para eliminar sufijos redundantes y capitalizar
# --------------------------------------------------------------
def limpiar_canonico(nombre):
    if not isinstance(nombre, str):
        return nombre
    # Quitar cualquier cosa después de ',' (por si quedaron restos)
    nombre = nombre.split(',')[0]
    # Capitalizar palabras importantes
    palabras = nombre.split()
    palabras = [p.capitalize() for p in palabras]
    return ' '.join(palabras)

# Aplicar limpieza a los valores del mapeo
mapeo = {k: limpiar_canonico(v) if v is not None else None for k, v in mapeo.items()}

# --------------------------------------------------------------
# 6. Aplicar el mapeo a la columna 'plantel_corregido' del DataFrame original
# --------------------------------------------------------------
df['plantel_estandarizado'] = df['plantel_corregido'].map(mapeo)

# --------------------------------------------------------------
# 7. Ver resultados
# --------------------------------------------------------------
print(f"Valores únicos originales: {df['plantel_corregido'].nunique()}")
print(f"Valores únicos después de fuzzy + mapeo: {df['plantel_estandarizado'].nunique()}")

# Guardar resultado
df.to_parquet("t_estandarizado_fuzzy.parquet", index=False)

# Guardar el mapeo usado
pd.DataFrame(list(mapeo.items()), columns=['original', 'estandarizado']).to_csv("mapeo_fuzzy.csv", index=False, encoding='utf-8')
```
